```python
import pandas as pd
import numpy as np
from random import *
import math
```

```python
s_weather = pd.read_csv('./data/weather0701.csv', encoding='euc-kr')
s_weather
```

```python
def cal(a,b):
  noise = 0
  d_list = []
  if (a<b):
    c = (b-a)/5
    for i in range(360):
      noise = uniform(-c,c)
      d_list.append(uniform(a,b)+noise)
  elif (a>b):
    c = (a-b)/5
    for i in range(360):
      noise = uniform(-c,c)
      d_list.append(uniform(a,b)+noise)
  else:
    for i in range(360):
      noise = uniform(-0.1,0.1)
      d_list.append(a+noise)
  
  return d_list
```

```python
s_weather = s_weather.fillna(0)
s_weather
```

```python
s_temp = s_weather['temperature']
s_temp
```

```python
s_Temp = []

for i in range(1464):
  s_Temp.append(cal(s_temp[i], s_temp[i+1]))
```

```python
print(s_Temp[30])
len(s_Temp[0])
```

```python
s_tempList = []

for i in s_Temp:
  s_tempList = s_tempList + i

len(s_tempList)
```

```python
s_df1 = pd.DataFrame(s_tempList)
```

```python
s_df1.to_csv('s_temperature.csv')
```

```python
def cal2(a,b):
  noise = 0
  d_list = []
  if (a<b):
    c = (b-a)/5
    for i in range(360):
      noise = uniform(-c,c)
      d_list.append(uniform(a,b)+noise)
  elif (a>b):
    c = (a-b)/5
    for i in range(360):
      noise = uniform(-c,c)
      d_list.append(uniform(a,b)+noise)
  else:
    if (a==0):
      for i in range(360):
        noise = 0
        d_list.append(a+noise)
    else:
      for i in range(360):
        noise = uniform(-0.01,0.01)
        d_list.append(a+noise)
  
  return d_list
```

```python
s_prec = s_weather['precipitation']
s_prec
```

```python
s_precList = []

for i in range(1464):
  s_precList.append(cal2(s_prec[i], s_prec[i+1]))

s_precList
```
